```python
import pandas as pd
import numpy as np
import random
import os
from datetime import datetime, timedelta

random.seed(42)
np.random.seed(42)

# Diretório de saída dinâmico (mesmo diretório do notebook)
output_dir = os.path.join(os.getcwd(), "petroquimica_dataset")
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# Dados auxiliares realistas para petroquímica brasileira
# ============================================================

# Cidades fictícias geradas com Faker (pt_BR)
estados_cidades = {
    "SP": ["Monte Teodoro", "Lagoa do Sul", "Várzea Arlindo", "Augusto do Norte", "Fagundes Verde", "Nova Tibério"],
    "BA": ["Lagoa das Águas", "Campo Raimundo", "Vila Henrique", "Baltazar do Cerrado"],
    "RS": ["Lagoa Augusto", "Santa Grande", "Cachoeira do Rio", "Arlindo Grande"],
    "RJ": ["Lagoa Branca", "Vila da Colina", "Santa Arlindo", "Ribeirão do Rio"],
    "AL": ["Arlindo do Sul", "Morro Joaquim"],
    "PE": ["Córrego Fagundes", "Serra da Colina", "Recanto Geraldo"],
    "MG": ["Nova Joaquim", "Várzea Bonita", "Floriano das Palmeiras"],
    "PR": ["Ribeirão das Águas", "Tibério Bonita", "Tibério do Sul"],
    "SC": ["Barra das Águas", "Nova da Mata"],
    "MT": ["Ponta Aparecida", "Várzea do Norte"],
}

# Unidades industriais fictícias
unidades_industriais = [
    "Polo Petroquímico de Lagoa das Águas",
    "Polo Petroquímico de Santa Grande",
    "Complexo Industrial de Monte Teodoro",
    "Unidade Lagoa Branca",
    "Polo Petroquímico de Fagundes Verde",
    "Unidade Industrial Várzea Arlindo",
    "Complexo Petroquímico de Córrego Fagundes (SUAPE)",
    "Unidade Industrial Ribeirão das Águas",
]

areas_responsaveis = [
    "Manutenção Industrial",
    "Engenharia de Processos",
    "Segurança e Meio Ambiente (SMS)",
    "Operações de Planta",
    "Logística e Suprimentos",
    "Qualidade e Laboratório",
    "Utilidades e Energia",
    "Planejamento e Controle de Manutenção (PCM)",
    "Projetos e Expansão",
    "Automação e Instrumentação",
]

categorias_fornecedor = [
    "Serviços de Manutenção Industrial",
    "Equipamentos e Peças",
    "Materiais de Consumo",
    "Serviços de Engenharia",
    "Transporte e Logística",
    "Serviços de Inspeção e Ensaios",
    "Produtos Químicos e Catalisadores",
    "Instrumentação e Automação",
    "Construção e Montagem",
    "Serviços Ambientais",
    "Serviços de Calibração",
    "Revestimentos e Isolamentos",
]

# Nomes ficticios de fornecedores gerados com Faker (pt_BR)
fornecedores_nomes = [
    "Abreu Manutencao",
    "Abreu Manutencao do Brasil",
    "Albuquerque Equipamentos",
    "Almeida Soldagem & Cia",
    "Alves Engenharia Ltda",
    "Andrade Dias",
    "Aparecida Mecanica do Brasil",
    "Aragão & Cia",
    "Aragão Mecanica",
    "Araújo Manutencao",
    "Araújo Moraes Automacao",
    "Araújo da Costa Equipamentos Ltda",
    "Azevedo Equipamentos & Cia",
    "Azevedo Isolamentos S.A.",
    "Azevedo Peixoto Servicos & Cia",
    "Barbosa Macedo",
    "Barros Ltda",
    "Borges S.A.",
    "Brito Brito Instrumentacao",
    "Brito Garcia Inspecao S.A.",
    "Caldeira Soldagem Industrial",
    "Camargo Inspecao S.A.",
    "Campos Tubulacoes",
    "Cardoso Ribeiro Logistica Industrial",
    "Carvalho Instrumentacao",
    "Carvalho Moraes Engenharia do Brasil",
    "Carvalho Sousa Servicos",
    "Casa Grande Instrumentacao",
    "Cassiano Calibracao & Cia",
    "Cassiano Equipamentos",
    "Cassiano Freitas Servicos S.A.",
    "Cassiano Isolamentos Ltda",
    "Castro S.A.",
    "Cirino Ambiental",
    "Correia Correia Industrial",
    "Câmara Azevedo",
    "Câmara Industrial",
    "Farias Equipamentos",
    "Ferreira Ltda",
    "Fogaça Industrial",
    "Fonseca Soldagem",
    "Garcia Industrial",
    "Garcia Montagens Ltda",
    "Gomes Inspecao",
    "Guerra Araújo",
    "Guerra Campos Logistica do Brasil",
    "Jesus Mendonça",
    "Jesus Quimica",
    "Leão Industrial",
    "Lopes Valvulas & Cia",
    "Marques Abreu",
    "Martins Garcia S.A.",
    "Martins Soldagem",
    "Montenegro Caldeira Industrial",
    "Montenegro Inspecao",
    "Montenegro Quimica & Cia",
    "Nascimento Equipamentos Ltda",
    "Nascimento Siqueira",
    "Nogueira Cardoso Equipamentos S.A.",
    "Novaes Ltda",
    "Nunes Silva Calibracao",
    "Oliveira Bombas do Brasil",
    "Pacheco Ltda",
    "Pacheco Peixoto S.A.",
    "Pastor Câmara",
    "Pastor Manutencao Ltda",
    "Pastor Soldagem",
    "Peixoto Equipamentos",
    "Pereira S.A.",
    "Pires Industrial",
    "Porto Industrial",
    "Rezende Monteiro Industrial Ltda",
    "Rios Silva Logistica",
    "Rios da Mata Industrial",
    "Rocha Quimica",
    "Rodrigues Ambiental",
    "Rodrigues Duarte S.A.",
    "Rodrigues Montagens",
    "Sampaio Calibracao do Brasil",
    "Sampaio Lima Quimica",
    "Santos Automacao Industrial",
    "Santos Azevedo Isolamentos do Brasil",
    "Silva das Neves Quimica",
    "Siqueira S.A.",
    "Souza & Cia",
    "Souza Novais Calibracao",
    "Sá Inspecao",
    "Teixeira Fernandes do Brasil",
    "Vargas Equipamentos",
    "Vargas da Rocha Logistica Ltda",
    "Vieira Rodrigues Automacao do Brasil",
    "da Conceicao Logistica Industrial",
    "da Conceicao Montenegro Ambiental",
    "da Cruz Ambiental S.A.",
    "da Luz Bombas",
    "da Luz Goncalves S.A.",
    "da Luz Moura",
    "da Mota Rios Industrial",
    "da Paz Equipamentos & Cia",
    "da Rocha Cunha Isolamentos do Brasil",
]

tipos_contrato = [
    "Contratação de Serviços de Manutenção",
    "Fornecimento de Materiais e Equipamentos",
    "Contratação de Serviços de Inspeção",
    "Serviços de Engenharia e Projetos",
    "Contratação de Serviços Ambientais",
    "Fornecimento de Produtos Químicos",
    "Serviços de Construção e Montagem",
    "Serviços de Automação e Instrumentação",
    "Contratação de Transporte e Logística",
    "Serviços de Calibração e Ensaios",
]

escopos_servico = [
    "Manutenção preventiva e corretiva de trocadores de calor e torres de resfriamento",
    "Fornecimento de válvulas de controle e segurança para unidades de polimerização",
    "Inspeção e ensaios não destrutivos (END) em vasos de pressão e tubulações",
    "Projeto e instalação de sistema de automação para planta de polietileno",
    "Tratamento de efluentes industriais e gestão de resíduos sólidos",
    "Fornecimento de catalisadores Ziegler-Natta para produção de polipropileno",
    "Montagem eletromecânica para expansão da unidade de PVC",
    "Calibração de instrumentos de medição de vazão, pressão e temperatura",
    "Transporte rodoviário de resinas termoplásticas em big bags e containers",
    "Manutenção de compressores e sistemas de refrigeração industrial",
    "Revestimento anticorrosivo de equipamentos e estruturas metálicas",
    "Isolamento térmico de tubulações de vapor e linhas de processo",
    "Limpeza química e hidrojateamento de permutadores e reatores",
    "Fornecimento de juntas, gaxetas e vedações para equipamentos rotativos",
    "Serviço de soldagem especializada (TIG/MIG) em aços inox e ligas especiais",
    "Instalação e comissionamento de analisadores de processo online",
    "Manutenção de sistemas elétricos e subestações de média tensão",
    "Fornecimento de bombas centrífugas e volumétricas para processo",
    "Consultoria em segurança de processos (HAZOP, análise de riscos)",
    "Serviços de andaime e acesso para paradas programadas de manutenção",
    "Inspeção baseada em risco (IBR) de equipamentos estáticos",
    "Fornecimento de filtros industriais para sistemas de ar e água",
    "Manutenção preditiva com análise de vibração e termografia",
    "Fornecimento de gases industriais (nitrogênio, oxigênio, hidrogênio)",
    "Projeto e fabricação de vasos de pressão conforme NR-13",
]

procedimentos_referencia = [
    "PROC-SMS-001_Permissao_Trabalho_Quente.pdf",
    "PROC-SMS-002_Bloqueio_Etiquetagem_LOTO.pdf",
    "PROC-SMS-003_Trabalho_em_Altura.pdf",
    "PROC-SMS-004_Espaco_Confinado.pdf",
    "PROC-SMS-005_Manuseio_Produtos_Quimicos.pdf",
    "PROC-MNT-001_Plano_Manutencao_Preventiva.pdf",
    "PROC-MNT-002_Procedimento_Parada_Programada.pdf",
    "PROC-MNT-003_Gestao_Sobressalentes.pdf",
    "PROC-QLD-001_Controle_Qualidade_Resinas.pdf",
    "PROC-QLD-002_Especificacao_Tecnica_Materiais.pdf",
    "PROC-ENG-001_Gestao_Mudancas_MOC.pdf",
    "PROC-ENG-002_Comissionamento_Startup.pdf",
    "PROC-AMB-001_Gestao_Residuos_Industriais.pdf",
    "PROC-AMB-002_Monitoramento_Emissoes.pdf",
    "PROC-LOG-001_Transporte_Produtos_Perigosos.pdf",
]

status_contrato = ["Ativo", "Ativo", "Ativo", "Ativo", "Encerrado", "Encerrado", "Suspenso", "Em Renovação", "Em Aprovação"]
status_ordem = ["Concluída", "Concluída", "Concluída", "Em Andamento", "Em Andamento", "Pendente", "Cancelada"]
prioridades = ["Baixa", "Média", "Média", "Alta", "Alta", "Crítica"]


def gerar_cnpj():
    return f"{random.randint(10,99)}.{random.randint(100,999)}.{random.randint(100,999)}/0001-{random.randint(10,99)}"

def gerar_telefone():
    ddd = random.choice([11, 13, 19, 21, 31, 41, 51, 71, 82, 81])
    return f"({ddd}) {random.randint(90000,99999)}-{random.randint(1000,9999)}"

# ============================================================
# TABELA 1: FORNECEDORES (~100 linhas)
# ============================================================
print("Gerando tabela de fornecedores...")

fornecedores_data = []
for i in range(1, 101):
    estado = random.choice(list(estados_cidades.keys()))
    cidade = random.choice(estados_cidades[estado])
    nome = fornecedores_nomes[i - 1] if i <= len(fornecedores_nomes) else f"Fornecedor Industrial {i}"
    
    fornecedores_data.append({
        "id_fornecedor": f"FORN-{i:04d}",
        "razao_social": nome,
        "cnpj": gerar_cnpj(),
        "categoria": random.choice(categorias_fornecedor),
        "cidade": cidade,
        "estado": estado,
        "telefone": gerar_telefone(),
        "email": nome.lower().replace(" ", ".").replace("/", "").replace("+", "").replace("'", "").replace("&", "e")[:30] + "@empresa.com.br",
        "status_cadastro": random.choices(["Ativo", "Inativo", "Bloqueado"], weights=[80, 15, 5])[0],
        "data_cadastro": (datetime(2018, 1, 1) + timedelta(days=random.randint(0, 2500))).strftime("%Y-%m-%d"),
        "avaliacao_desempenho": round(random.uniform(5.0, 10.0), 1),
    })

df_fornecedores = pd.DataFrame(fornecedores_data)
print(f"Fornecedores: {len(df_fornecedores)} linhas")

# ============================================================
# TABELA 2: CONTRATOS (1000 linhas)
# ============================================================
print("Gerando tabela de contratos...")

contratos_data = []
for i in range(1, 1001):
    id_forn = random.choice(df_fornecedores["id_fornecedor"].tolist())
    tipo = random.choice(tipos_contrato)
    escopo = random.choice(escopos_servico)
    unidade = random.choice(unidades_industriais)
    area = random.choice(areas_responsaveis)
    data_inicio = datetime(2020, 1, 1) + timedelta(days=random.randint(0, 2000))
    duracao_dias = random.choice([90, 180, 365, 545, 730, 1095])
    data_fim = data_inicio + timedelta(days=duracao_dias)
    
    # Valor proporcional ao tipo
    if "Materiais" in tipo or "Produtos Químicos" in tipo:
        valor = round(random.uniform(50_000, 5_000_000), 2)
    elif "Engenharia" in tipo or "Construção" in tipo:
        valor = round(random.uniform(200_000, 15_000_000), 2)
    elif "Transporte" in tipo:
        valor = round(random.uniform(100_000, 3_000_000), 2)
    else:
        valor = round(random.uniform(80_000, 8_000_000), 2)
    
    # Documentos de referência (contrato PDF + procedimento)
    num_contrato = f"CT-{data_inicio.year}-{i:05d}"
    doc_contrato = f"CONTRATO_{num_contrato}.pdf"
    doc_procedimento = random.choice(procedimentos_referencia)
    
    contratos_data.append({
        "id_contrato": f"CTR-{i:05d}",
        "numero_contrato": num_contrato,
        "id_fornecedor": id_forn,
        "tipo_contrato": tipo,
        "descricao_escopo": escopo,
        "valor_total": valor,
        "moeda": "BRL",
        "data_inicio": data_inicio.strftime("%Y-%m-%d"),
        "data_fim": data_fim.strftime("%Y-%m-%d"),
        "status": random.choice(status_contrato),
        "unidade_industrial": unidade,
        "area_responsavel": area,
        "documento_contrato": doc_contrato,
        "procedimento_referencia": doc_procedimento,
    })

df_contratos = pd.DataFrame(contratos_data)
print(f"Contratos: {len(df_contratos)} linhas")

# ============================================================
# TABELA 3: ORDENS DE SERVIÇO (~500 linhas)
# ============================================================
print("Gerando tabela de ordens de serviço...")

ordens_data = []
contratos_ativos = df_contratos[df_contratos["status"].isin(["Ativo", "Em Renovação"])]["id_contrato"].tolist()
if len(contratos_ativos) < 50:
    contratos_ativos = df_contratos["id_contrato"].tolist()

for i in range(1, 501):
    id_ctr = random.choice(contratos_ativos)
    contrato_info = df_contratos[df_contratos["id_contrato"] == id_ctr].iloc[0]
    
    data_emissao = datetime.strptime(contrato_info["data_inicio"], "%Y-%m-%d") + timedelta(days=random.randint(0, 365))
    prazo_dias = random.choice([7, 14, 30, 45, 60, 90])
    data_prevista = data_emissao + timedelta(days=prazo_dias)
    status_os = random.choice(status_ordem)
    data_conclusao = (data_prevista + timedelta(days=random.randint(-10, 30))).strftime("%Y-%m-%d") if status_os == "Concluída" else None
    
    valor_os = round(contrato_info["valor_total"] * random.uniform(0.01, 0.15), 2)
    
    descricoes_os = [
        f"Execução de {contrato_info['descricao_escopo'][:60].lower()} - Etapa {random.randint(1,5)}",
        f"Mobilização de equipe para {contrato_info['tipo_contrato'].lower()} na {contrato_info['unidade_industrial']}",
        f"Fornecimento parcial conforme contrato {contrato_info['numero_contrato']}",
        f"Atendimento emergencial - {contrato_info['area_responsavel']}",
        f"Medição mensal #{random.randint(1,12)} - {contrato_info['tipo_contrato']}",
    ]
    
    ordens_data.append({
        "id_ordem": f"OS-{i:05d}",
        "numero_ordem": f"OS-{data_emissao.year}-{random.randint(10000,99999)}",
        "id_contrato": id_ctr,
        "id_fornecedor": contrato_info["id_fornecedor"],
        "tipo_ordem": random.choice(["Serviço", "Fornecimento", "Manutenção Corretiva", "Manutenção Preventiva", "Emergencial"]),
        "descricao": random.choice(descricoes_os),
        "valor_ordem": valor_os,
        "data_emissao": data_emissao.strftime("%Y-%m-%d"),
        "data_prevista_conclusao": data_prevista.strftime("%Y-%m-%d"),
        "data_conclusao": data_conclusao,
        "status": status_os,
        "prioridade": random.choice(prioridades),
        "unidade_industrial": contrato_info["unidade_industrial"],
        "area_solicitante": contrato_info["area_responsavel"],
    })

df_ordens = pd.DataFrame(ordens_data)
print(f"Ordens de Serviço: {len(df_ordens)} linhas")

# ============================================================
# SALVAR CSVs no Workspace File System
# ============================================================
df_fornecedores.to_csv(f"{output_dir}/fornecedores.csv", index=False, encoding="utf-8-sig")
df_contratos.to_csv(f"{output_dir}/contratos.csv", index=False, encoding="utf-8-sig")
df_ordens.to_csv(f"{output_dir}/ordens_servico.csv", index=False, encoding="utf-8-sig")

print(f"\nCSVs salvos em: {output_dir}")
print(f"  - fornecedores.csv    ({len(df_fornecedores)} linhas, {len(df_fornecedores.columns)} colunas)")
print(f"  - contratos.csv       ({len(df_contratos)} linhas, {len(df_contratos.columns)} colunas)")
print(f"  - ordens_servico.csv  ({len(df_ordens)} linhas, {len(df_ordens.columns)} colunas)")

print("\n--- Amostra: Fornecedores ---")
display(df_fornecedores.head(5))
print("\n--- Amostra: Contratos ---")
display(df_contratos.head(5))
print("\n--- Amostra: Ordens de Serviço ---")
display(df_ordens.head(5))
```
